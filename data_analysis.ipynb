```python
import glob2
import pandas as pd
```

```python
# List out the files needed using the glob module
bike_data = glob2.glob("source_files/trip_data/*.csv")
bike_data
```

```python
# Import all the files and combine them into one dataframe
all_bike_data_df = pd.DataFrame()
for f in bike_data:
    df = pd.read_csv(f)
    all_bike_data_df = all_bike_data_df.append(df,ignore_index=True, sort=False)
```

```python
all_bike_data_df.head()
```

```python
all_bike_data_df.count()
```

```python
all_bike_data_df.to_csv("source_data/2018_bike_data.csv")
```
